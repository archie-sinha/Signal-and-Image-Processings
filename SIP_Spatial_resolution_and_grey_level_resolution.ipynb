```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
from google.colab.patches import cv2_imshow
```

```python
image= cv.imread('cameraman.png',0)
cv2_imshow(image)
image.shape
```

```python
#DOWN SAMPLING using slicing:

#Taking value from user
a= int(input("ENTER A DOWNSAMPLING NO. FOR ROWS: "))
b =int(input("ENTER A DOWNSAMPLING NO. FOR COLUMNS: "))

image1= image[::a, ::b]
cv2_imshow(image1)
image1.shape
```

```python
#UPSAMPLING:

m,n= image.shape
f=4
image2 = np.zeros((m, n), dtype=np.int)

# new size
for i in range(0, m-1, f):
    for j in range(0, n-1, f):
        image2[i, j] = image1[i//f][j//f]

# Nearest neighbour interpolation-Replication
# Replicating rows

for i in range(1, m-(f-1), f):
    for j in range(0, n-(f-1)):
        image2[i:i+(f-1), j] = image2[i-1, j]

# Replicating columns
for i in range(0, m-1):
    for j in range(1, n-1, f):
        image2[i, j:j+(f-1)] = image2[i, j-1]

# Plot the up sampled image
print('Up Sampled Image:')
plt.imshow(image2, cmap="gray")
image2.shape
```

```python
#Upsampling
#Create matrix of zeros to store the upsampled image
image_2 = np.zeros((m, n), dtype=np.int)
f=int(input("Enter the upsampling rate: "))
for i in range(0, m-1, f):
    for j in range(0, n-1, f):
        image_2[i,j] = image1[i//f][j//f]
print(image_2)
# Replicating rows
for i in range(1, m-(f-1), f):
    for j in range(0, n-(f-1)):
        image_2[i:i+(f-1), j] = image_2[i-1, j]
print(image_2)
# Replicating columns
for i in range(0, m-1):
    for j in range(1, n-1, f):
        image_2[i, j:j+(f-1)] = image_2[i, j-1]
print(image_2)
plt.imshow(image_2, cmap="gray")
```

# Apply sampling resolution on the test image to obtain the new image.

i. Calculate the compression ratio between the test image and new image

ii. Comment on subjective quality of the image for various down sampling rates

```python
#Gray level
height, width= image.shape
gray_image = np.zeros((height, width), dtype=np.uint8)
a = int(input("Enter bits of new image: "))
b=2**a
for i in range(height):
    for j in range(width):
        gray_image[i, j] = (b-1)/(2**8-1)*image[i, j]
print(gray_image)
plt.imshow(gray_image,cmap='gray')
plt.show()
plt.imshow(gray_image)
plt.show()
```
