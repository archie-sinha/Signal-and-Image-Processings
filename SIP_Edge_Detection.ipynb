```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
image1=cv.imread('/content/Fig1016(a)(building_original).tif',0)
plt.imshow(image1,cmap='gray')
```

```python
[m,n]=image1.shape
print(m,n)
```

Horizontal Edge:

```python
mask=np.array([[-1,-2,-1],[0,0,0],[1,2,1]])
```

```python
print(mask)
```

```python
print(image1)
```

```python
img1=np.zeros([m,n],dtype=int)
for i in range(1,m-1):
  for j in range(1,n-1):
    temp=image1[i-1:i+2,j-1:j+2]
    constant=mask*temp
    constant=np.sum(constant)
    img1[i][j]=constant

plt.imshow(img1,cmap='gray',vmax=255,vmin=0)
```

Vertical Edge:

```python
mask2=np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
```

```python
print(mask2)
```

```python
img2=np.zeros([m,n],dtype=int)
for i in range(1,m-1):
  for j in range(1,n-1):
    temp=image1[i-1:i+2,j-1:j+2]
    constant=mask2*temp
    constant=np.sum(constant)
    img2[i][j]=constant

plt.imshow(img2,cmap='gray',vmax=255,vmin=0)
```
