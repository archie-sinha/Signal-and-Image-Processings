```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
frequency = int(input("Enter Frequency: "))
f= frequency
```

```python
t=np.arange(0,1,0.1)
# print(t)
x= np.sin(2*np.pi*f*t)
y= np.cos(2*np.pi*f*t)
```

```python
plt.subplot(2,2,1)  #2 rows, 2columns, 1st Part
plt.plot(t,x)
plt.subplot(2,2,2)
plt.stem(t,x)

plt.subplot(2,2,3)  #2 rows, 2columns, 3rd Part
plt.plot(t,y)
plt.subplot(2,2,4)
plt.stem(t,y)
```

UNIT STEP

u(t)= {1, t<=5
              0, otherwise

```python
t=np.arange(0,10,1)
f=1
u=[]
i=0
for i in t:
  if(t[i]<=5):
    u.append(1)
  else:
        u.append(0)

plt.subplot(1,2,1)
plt.plot(t,u)
plt.subplot(1,2,2)
plt.stem(t,u)
```

EXPONENTIAL

n=0 to 10

x(n)= a^n

Case 1: -1<=a<0    (a=-0.5)

Case 2: a<=-1      (a=-2)

Case 3: a>1        (a=2)

Case 4: 0<a<=1     (a=0.5)

```python
n=np.arange(0,10,1)
x=[]
a= int(input("Enter a number: "))
for i in n:
  temp=a**i
  x.append(temp)


plt.subplot(1,2,1)
plt.plot(n,x)
plt.subplot(1,2,2)
plt.stem(n,x)
```

```python
n=np.arange(0,10,1)
a= float(input("Enter a number: "))
x= a**n
plt.subplot(1,2,1)
plt.plot(n,x)
plt.subplot(1,2,2)
plt.stem(n,x)
```

TIME SHIFTING OPERATIONS

a. Time Advance

b. Time Delay

c. Time Reversal

d. Time Reversal & Delay

e. Time Reversal & Advance

```python
# TIME DELAY
n= np.arange(-2,4,1)
x= n
y= n-3

plt.subplot(2,2,1)
plt.plot(n,x)
plt.subplot(2,2,2)
plt.stem(n,x)
plt.subplot(2,2,3)
plt.plot(n,y)
plt.subplot(2,2,4)
plt.stem(n,y)
```

```python
# TIME REVERSE
n= np.arange(2,8,1)
x= n
y= -x

plt.subplot(2,2,1)
plt.plot(n,x)
plt.subplot(2,2,2)
plt.stem(n,x)
plt.subplot(2,2,3)
plt.plot(n,y)
plt.subplot(2,2,4)
plt.stem(n,y)
```

```python
# TIME REVERSE & ADVANCE
n= np.arange(2,8,1)
x= n
y= -n-1

plt.subplot(2,2,1)
plt.plot(n,x)
plt.subplot(2,2,2)
plt.stem(n,x)
plt.subplot(2,2,3)
plt.plot(n,y)
plt.subplot(2,2,4)
plt.stem(n,y)
```

TIME SCALING OPERATIONS

a. Down Sampling

b. Up Scaling

```python
#Down Sampling

x={0:1, 1:2, 2:3, 3:4, 4:5}
a=[]
b=[]
for i in range (0, len(x), 1):
  if (i*2)<=4:
    a.append(i)
    b.append(x[2*i])

plt.stem(a,b)
```

```python
#Up Sampling
n= np.arange(-2,6,1)
x= n
y=n/2

plt.subplot(2,2,1)
plt.plot(n,x)
plt.subplot(2,2,2)
plt.stem(n,x)
plt.subplot(2,2,3)
plt.plot(n,y)
plt.subplot(2,2,4)
plt.stem(n,y)
```

SCALAR OPERATIONS

a. Addition

b. Subtraction

c. Multiplication

```python
# Addition
x = np.arange(0,10,0.01)
x2 = np.arange(0,20,0.02)

sin1 = np.sin(x)
sin2 = np.sin(x2)

x2 /= 2

sin3 = sin1+sin2

plt.plot(x,sin1, color="pink")
plt.show()
plt.plot(x,sin2,color="purple")
plt.show()
plt.plot(x,sin3, color="magenta")
plt.show()
```

```python
# Subtraction
x = np.arange(0,10,0.01)
x2 = np.arange(0,20,0.02)

sin1 = np.sin(x)
sin2 = np.sin(x2)

x2 /= 2

sin3 = sin1-sin2

plt.plot(x,sin1)
plt.show()
plt.plot(x,sin2)
plt.show()
plt.plot(x,sin3)
plt.show()
```

```python
# Multiplication
x = np.arange(0,10,0.01)
x2 = np.arange(0,20,0.02)

sin1 = np.sin(x)
sin2 = np.sin(x2)

x2 /= 2

sin3 = sin1*sin2

plt.plot(x,sin1)
plt.show()
plt.plot(x,sin2)
plt.show()
plt.plot(x,sin3)
plt.show()
```
