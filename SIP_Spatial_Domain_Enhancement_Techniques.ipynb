ii.	To write a program in PYTHON to obtain thresholding of an image

```python
image_1 = np.zeros((m,n), dtype=np.int)

for i in range(0, m):
    for j in range(0,n):
        try:
          if image[i][j]>150:
            image_1[i][j] = 255
        except IndexError:
            pass

plt.imshow(image_1, cmap="gray")
```

iii.	To write a program in PYTHON to obtain grey level slicing of an image without background

```python
image_2 = np.zeros((m,n), dtype=np.int)

for i in range(0, m):
    for j in range(0,n):
        try:
          if image[i][j]<100:
            image_2[i][j] = 0
          elif 100<image[i][j]<150:
            image_2[i][j] = 125
          elif image[i][j]>150:
            image_2[i][j] = 0
        except IndexError:
            pass

plt.imshow(image_2, cmap="gray")
```

iv.	To write a program in PYTHON to obtain grey level slicing of an image with background

```python
image_3 = np.zeros((m,n), dtype=np.int)

for i in range(0, m):
    for j in range(0,n):
        try:
          if image[i][j]<100:
            image_3[i][j] = image[i][j]
          elif 100<image[i][j]<150:
            image_3[i][j] = 150
          elif image[i][j]>150:
            image_3[i][j] = image[i][j]
        except IndexError:
            pass

plt.imshow(image_3, cmap="gray")
```
