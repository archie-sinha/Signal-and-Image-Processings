```python
SE=np.ones((11,11), dtype=int)
```

```python
SE
```

```python
constant=5
imgE=np.zeros([m,n],dtype=int)
for i in range(5, m-5):
    for j in range(5,n-5):
        temp= img[i-5:i+6,j-5:j+6]
        constant= temp*SE

        imgE[i,j]= np.min(constant)

plt.imshow(imgE,cmap="gray")
```

```python
constant=5
imgD=np.zeros([m,n],dtype=int)
for i in range(5, m-5):
    for j in range(5,n-5):
        temp= img[i-5:i+6,j-5:j+6]
        constant= temp*SE
        imgD[i,j]= np.max(constant)

plt.imshow(imgD,cmap="gray")
```
